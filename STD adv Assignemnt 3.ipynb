```python
# Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.
```

```python
from math import sqrt
from scipy.stats import norm
```

```python
def estimate_population_mean(sample_mean, std_deviation, sample_size, confidence_level):
    z_score = norm.ppf(1 - ((1 - confidence_level) / 2))
    return sample_mean + (z_score * (std_deviation / sqrt(sample_size)))
```

```python
# Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis 
#that the population means are equal.
```

```python
import numpy as np
from scipy.stats import ttest_ind

def two_sample_t_test(sample1, sample2, alpha=0.05):
    # Perform two-sample t-test
    t_statistic, p_value = ttest_ind(sample1, sample2)

    # Check the p-value against the significance level (alpha)
    if p_value < alpha:
        print("Reject the null hypothesis: The population means are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The population means are not significantly different.")

# Example usage:
sample1 = np.array([10, 12, 15, 20, 18])
sample2 = np.array([13, 14, 16, 17, 19])

two_sample_t_test(sample1, sample2)
```

```python
# Q12: What is t-statistic? State the formula for t-statistic.
```

```python
# The t-statistic is a measure used in hypothesis testing to assess whether the sample mean differs significantly from the hypothesized population mean.
It quantifies the difference between the sample mean and the population mean in terms of standard errors.

Formula for t-statistic:
t = (x̄ - μ) / (s / √n)

Where:
t = t-statistic
x̄ = Sample mean
μ = Population mean (hypothesized value under the null hypothesis)
s = Sample standard deviation
n = Sample size
```

```python
#Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random #sample of 50 days and find the sample mean revenue to be $500 
with a standard deviation of $50. #Estimate the population mean revenue with a 95% confidence interval.
```

```python
import numpy as np
from scipy.stats import t

# Given data
sample_mean = 500
sample_std_dev = 50
sample_size = 50
confidence_level = 0.95

# Calculate the t-critical value
t_critical = t.ppf((1 + confidence_level) / 2, df=sample_size - 1)

# Calculate the Margin of Error
margin_of_error = t_critical * (sample_std_dev / np.sqrt(sample_size))

# Calculate the Confidence Interval
confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

# Print the result
print(f"The 95% Confidence Interval for the population mean revenue is: ${confidence_interval_lower:.2f} to ${confidence_interval_upper:.2f}")
```

```python
# Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a clinical trial with 100 patients and find that the sample mean decrease in
# blood pressure is 8 mmHg with a standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.
```

```python
import numpy as np
from scipy.stats import t

# Given data
sample_mean = 8
hypothesized_mean = 10
sample_std_dev = 3
sample_size = 100
significance_level = 0.05

# Calculate the t-statistic
t_statistic = (sample_mean - hypothesized_mean) / (sample_std_dev / np.sqrt(sample_size))

# Calculate the degrees of freedom
df = sample_size - 1

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - t.cdf(abs(t_statistic), df))

# Compare p-value with significance level
if p_value < significance_level:
    print("Reject the null hypothesis: The drug decreases blood pressure significantly.")
else:
    print("Fail to reject the null hypothesis: The drug does not significantly decrease blood pressure.")
```

```python
#Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 =
#30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean
#score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two
#groups are equal with a significance level of 0.01.
```

```python
The formula for the two-sample t-test is:

t = (mean1 - mean2) / √((s1^2 / n1) + (s2^2 / n2))
```

```python
import numpy as np
from scipy.stats import t

# Group 1 data
mean1 = 80
std_dev1 = 10
sample_size1 = 30

# Group 2 data
mean2 = 75
std_dev2 = 8
sample_size2 = 40

# Significance level
significance_level = 0.01

# Calculate the pooled standard deviation (pooled variance)
pooled_std_dev = np.sqrt(((std_dev1**2 * (sample_size1 - 1)) + (std_dev2**2 * (sample_size2 - 1))) / (sample_size1 + sample_size2 - 2))

# Calculate the t-statistic
t_statistic = (mean1 - mean2) / np.sqrt((std_dev1**2 / sample_size1) + (std_dev2**2 / sample_size2))

# Calculate the degrees of freedom
df = sample_size1 + sample_size2 - 2

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - t.cdf(abs(t_statistic), df))

# Compare p-value with significance level
if p_value < significance_level:
    print("Reject the null hypothesis: The population means for the two groups are not equal.")
else:
    print("Fail to reject the null hypothesis: The population means for the two groups are equal.")
```

```python
# Q17. A marketing company wants to estimate the average number of ads watched by viewers during a TV
program. They take a random sample of 50 viewers and find that the sample mean is 4 with a standard
deviation of 1.5. Estimate the population mean with a 99% confidence interval.
```
